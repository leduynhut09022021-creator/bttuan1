```python
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import folium
geolocator = Nominatim(user_agent="geo_app")
ds_dd = [
    "Ben Thanh Market, Ho Chi Minh City",
    "Independence Palace, Ho Chi Minh City",
    "Notre Dame Cathedral, Ho Chi Minh City",
    "War Remnants Museum, Ho Chi Minh City",
    "Nguyen Hue Walking Street, Ho Chi Minh City",
    "Bitexco Tower, Ho Chi Minh City",
    "Saigon Zoo, Ho Chi Minh City",
    "Tan Dinh Church, Ho Chi Minh City",
    "Landmark 81, Ho Chi Minh City",
    "GEM Center, Ho Chi Minh City"
]
diem_trung_tam = "International Square, Ho Chi Minh city"
center = geolocator.geocode(diem_trung_tam)
if center:
    toa_do_trung_tam = (center.latitude, center.longitude)
    print(f" Tọa độ trung tâm: {toa_do_trung_tam}")
danh_sach_toa_do = []

for dia_chi in ds_dd:
  loc = geolocator.geocode(dia_chi)
  if loc:
          toa_do = (loc.latitude, loc.longitude)
          kc = geodesic(toa_do_trung_tam,toa_do).km
          danh_sach_toa_do.append({
                "Ten": dia_chi.split(',')[0],
                "Toa_Do": toa_do,
                "Khoang_Cach": round(kc, 2)
            })
m = folium.Map(location=toa_do_trung_tam, zoom_start=13)
folium.Marker(location=toa_do, popup="dia_chi", tooltip="Click vào").add_to(m)
folium.PolyLine([[toa_do_trung_tam], [toa_do]], color="blue").add_to(m)
m





```
